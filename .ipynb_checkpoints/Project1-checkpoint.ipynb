```python
import networkx as nx
seed=1000           # seed the graph for reproducibility, you should be doing this  
G= nx.gnp_random_graph (200, .01, seed=seed )       # here we create a random binomial graph with 20 nodes and an average (expected) connectivity of 10*.3= 3.
print ( G.nodes() )
```

```python
print(G.edges())
```
